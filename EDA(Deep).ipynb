# Positive Words Count

```python
df_pos = pd.DataFrame(dataset[dataset['sentiment'] == 'positive']['text_ppd'], columns=['text_ppd'])
```

```python
df_pos
```

```python
vocab = set()
for index, row in df_pos.iterrows():
    text = row['text_ppd']
    tokens = text.split()
    for token in tokens:
        if len(token) >= 6:
            vocab.add(token)
```

```python
# create dictionary to store word counts
word_count = {}
for word in vocab:
    count = 0
    for row in df_pos['text_ppd']:
        if word in row:
            count += 1
    word_count[word] = count
```

```python
# remove a list of words from word_count dictionary
words_to_remove = ['zing','mazing','really','definite','deli', 'licious','food', 'dian', 'india', 'ever', 'place', 'chic', 'hick', 'chick', 'rest', 'rant', 'vice', 'thin', 'order', 'thing', 'every', 'able', 'back', 'dish', 'like', 'also']
for word in words_to_remove:
    if word in word_count:
        del word_count[word]
```

```python
# sort dictionary by word count in descending order
sorted_word_count_pos = dict(sorted(word_count.items(), key=lambda item: item[1], reverse=True))
```

```python
# get top 20 most frequent words
top_words = dict(sorted(sorted_word_count_pos.items(), key=lambda x: x[1], reverse=True)[:20])
```

```python
df_top_words = pd.DataFrame.from_dict(top_words, orient='index', columns=['count'])
df_top_words.index.name = 'word'

#print(df_top_words)
```

```python
df_top_words
```

```python
# plot the bar chart
ax = df_top_words.plot(kind='barh', figsize=(8,6), color='purple')
ax.set_title('Top 20 Words in Positive Sentiment Reviews')
ax.set_xlabel('Count')
ax.set_ylabel('Word')
ax.invert_yaxis()
plt.show()
```

# Negative Word Count

# Top 10 restaurants with minimum 50 reviews based on avg reviews. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# group the DataFrame by restaurant name and calculate the count and mean review scores
grouped = dataset.groupby('name')['review_stars'].agg(['count', 'mean'])

# filter out restaurants with less than 5 reviews and get the top 10 by mean review score
top_restaurants = grouped[grouped['count'] >= 100].nlargest(10, 'mean')

# create a color palette with 15 distinct colors
palette = sns.color_palette("husl", 15)

# plot the top 10 restaurants by mean review score
fig, ax = plt.subplots()
ax.barh(top_restaurants.index, top_restaurants['mean'], color=palette)
ax.invert_yaxis()  # invert the y-axis to show the top restaurant at the top
ax.set_xlabel('Average Review Score')
ax.set_ylabel('Restaurant')
ax.set_title('Top 10 Restaurants with at Least 100 Reviews')
plt.show()
```
